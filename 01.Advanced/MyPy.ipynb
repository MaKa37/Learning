# MyPy

`MyPy`란 파이썬 코드의 Type Hint를 분석하여 코드 실행 전 잠재적인 오류를 찾아내는 정적 분석 도구입니다.

파이썬의 유연한 동적 타이핑 특성을 유지하면서,

대규모 프로젝트에 발생할 수 있는 '타입 불일치' 문제를 방지하기 위해 설계 되었습니다.

MyPy를 활용 시 코드 작성 단계에서 즉각적인 피드백을 받을 수 있어,

런타임 `TypeError`가 발생하는 상황을 획기적으로 줄일 수 있습니다.

## MyPy 핵심 명령어 요약

## 점진적 채택(Gradual Typing)

점진적 채택이란 처음 로직을 구상할때는 타입 힌트 없이 `Any`를 사용하여 구상하고

중요도가 높거나 버그가 자주 발생하는 함수부터 점진적으로 구체적인 타입을 채택하고,

점차 `Any`를 줄여나가는 방식입니다.

최종적으로는 **엄격 모드(`--strict`)**를 사용하면 MyPy는 다음을 강제합니다.
- 모든 함수에 타입 힌트가 반드시 있어야 함.
- 타입을 알 수 없는 `Any` 사용을 최대한 제한함.
- 체크하지 않고 넘어가는 코드가 없도록 함.

### 🛠️ 실무 팁: 설정 파일 (`mypy.ini` 또는 `pyproject.toml`)

터미널에서 매번 --strict 같은 긴 옵션을 입력하는 것은 번거롭고 실수하기 쉽습니다.

그래서 실제 프로젝트에서는 설정 파일을 만들어 관리합니다.

예를 들어, pyproject.toml 파일에 다음과 같이 적어두면 단순히 mypy .만 입력해도 엄격한 검사가 실행됩니다.

```TOML
[tool.mypy]
python_version = "3.10"
strict = true
ignore_missing_imports = true
```

## 핵심 용어 설명

| **용어**                  | **설명**                                          | **MyPy에서의 역할**                         |
| ----------------------- | ----------------------------------------------- | -------------------------------------- |
| **Static Analysis** 🔍  | 코드를 **실행하지 않고** 소스 코드 자체를 분석하여 오류를 찾는 방법입니다.    | 실행 전 단계에서 타입 불일치나 논리적 오류를 미리 잡아냅니다.    |
| **Linting** 🧹          | 코드의 스타일, 잠재적 에러, 가독성 등을 체크하는 도구 또는 과정입니다.       | MyPy는 일종의 '타입 전용 린터'로서 코드의 품질을 높여줍니다.  |
| **Type Consistency** ⚖️ | 변수에 할당된 값의 타입과 선언된 타입이 **일관성** 있게 유지되는 상태입니다.   | 함수 인자나 반환값이 약속된 타입과 일치하는지 지속적으로 검사합니다. |
| **Strict Mode** 🚫      | MyPy의 검사 수준을 가장 높여, 아주 작은 타입 누락도 허용하지 않는 모드입니다. | 모든 곳에 타입 힌트를 강제하여 코드의 안정성을 극대화합니다.     |

## MyPy 설정

MyPy의 **설정 파일**은 프로젝트 전체에 동일한 정적 검사 규칙을 적용하여 협업 시 "누구 컴퓨터에서는 에러가 나고, 누구는 안 나는" 상황을 방지해 줍니다. 특히 **점진적 채택(Gradual Typing)**을 할 때는 처음부터 너무 엄격하게 설정하면 에러가 수천 개씩 쏟아질 수 있으므로, 설정을 통해 검사 강도를 조절하는 것이 핵심입니다. ⚙️

주로 `pyproject.toml` 또는 `mypy.ini` 파일을 사용하는데, 최신 파이썬 프로젝트에서는 다른 도구들과 설정을 통합할 수 있는 **`pyproject.toml`** 형식을 권장합니다.


### 🛠️ 핵심 구성 옵션 상세 설명

1. **`python_version`**: 프로젝트가 사용하는 파이썬 버전을 명시합니다. 버전에 따라 지원되는 타입 힌트 문법이 다르기 때문에 중요합니다. 🐍
    
2. **`strict = true`**: "엄격 모드"입니다. 이 옵션을 켜면 MyPy의 거의 모든 엄격한 검사 플래그가 한 번에 활성화됩니다. 신규 프로젝트라면 처음부터 켜는 것이 좋지만, 기존 프로젝트는 `false`로 시작하는 것이 좋습니다. 🚫
    
3. **`ignore_missing_imports`**: 타입 힌트 정보를 제공하지 않는 외부 라이브러리(third-party library)를 가져올 때 발생하는 에러를 무시합니다. 실무에서 아주 자주 쓰이는 옵션입니다. 🙈
    
4. **`disallow_untyped_defs`**: 타입 힌트가 없는 함수 정의를 금지합니다. 점진적 채택 시 이 옵션을 특정 모듈에만 적용하여 점차적으로 타입을 입힐 수 있습니다. ✍️
    
5. **`check_untyped_defs`**: 함수 내부에 타입 힌트가 없더라도, 그 안의 로직은 검사하도록 설정합니다. 최소한의 안전장치 역할을 합니다. 🛡️

### 📑 MyPy 설정 옵션 요약표

|**옵션명**|**기본값**|**설명**|**점진적 채택 시 팁**|
|---|---|---|---|
|`python_version`|(현재 실행 버전)|검사 대상 파이썬 버전 지정|프로젝트의 실제 실행 버전과 일치시킬 것|
|`strict`|`false`|모든 엄격 모드 플래그 활성화|처음엔 `false`, 점차 `true`로 전환 지향|
|`ignore_missing_imports`|`false`|타입 정보 없는 라이브러리 경고 무시|외부 라이브러리 에러가 너무 많을 때 `true` 설정|
|`disallow_untyped_defs`|`false`|타입 힌트가 없는 함수 선언 금지|새로 작성하는 폴더/파일에만 우선 적용 추천|
|`check_untyped_defs`|`false`|힌트 없는 함수 내부 로직 검사|타입 힌트를 적기 전 로직 검증용으로 유용|
|`warn_return_any`|`false`|`Any` 타입을 반환할 때 경고 발생|반환 값의 타입을 명확히 강제하고 싶을 때 사용|

### 🔍 직접 구성해 보기

아래는 `pyproject.toml` 파일의 일부입니다.

Ini, TOML

```TOML
[tool.mypy]
python_version = "3.10"
ignore_missing_imports = true
check_untyped_defs = true

[[tool.mypy.overrides]]
module = "legacy_code.*"
disallow_untyped_defs = false  # 옛날 코드 폴더는 힌트 없어도 허용
```

위 설정에서 **`[[tool.mypy.overrides]]`** 부분을 주목해 보세요.

특정 폴더(`legacy_code`)에만 다른 규칙을 적용하고 있습니다.
